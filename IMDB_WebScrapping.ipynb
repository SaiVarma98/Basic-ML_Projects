```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://www.imdb.com/chart/moviemeter/?sort=rk,asc&mode=simple&page=1").text #Add headers if it showing bad request
```

```python
soup=BeautifulSoup(page,"lxml")
```

```python
print(soup.prettify())
```

```python
## Get all poster images and make a list 
```

```python
posters=soup.find_all("td",class_="posterColumn")
title=soup.find_all("td",class_="titleColumn")
rating=soup.find_all("td",class_="ratingColumn imdbRating")
```

```python
image=[]
tt=[]
release_year=[]
IMDB_rating=[]

for i in range(len(rating)) :

    
        if rating[i].find("strong") :
           s=''
           image.append(posters[i].find("a").img.get("src"))
           tt.append(title[i].find("a").text)
           release_year.append(title[i].find("span",class_='secondaryInfo').string[1:5])    
           IMDB_rating.append(s.join([rating[i].find("strong").string]))
        
           
        
```

```python
print(len(image))
```

```python
data=pd.DataFrame(list(zip(image,tt,release_year,IMDB_rating)),columns =['Image', 'Title','Release_Year','Imdb_Rating'])
```
